```python
import os
import wget
import time
import requests
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

project_path = os.path.abspath(os.path.dirname('__file__'))
driver_path = os.path.join(project_path, "chromedriver")
driver = webdriver.Chrome(driver_path)
driver.get('https://search.damai.cn/search.htm?spm=a2oeg.home.category.ditem_1.591b23e1yDqddL&ctl=%E8%AF%9D%E5%89%A7%E6%AD%8C%E5%89%A7&order=1&cty=%E4%B8%8A%E6%B5%B7')


header = ['Picture','Title', 'Venue', 'Time', 'Price','Link']
data = []

for i in range(1,4):
    html = driver.page_source
    soup = BeautifulSoup(html)
    for item in soup.find_all("div", attrs = {"class":"items"}):
        Picture = item.find("img", attrs = {"alt":"项目图片"}).get("data-src")
        Title = item.find ("div", attrs = {"class":"items__txt__title"}).text.strip()
        Venue1 = item.find ("a", attrs = {"class":"items__txt__venue__icon"})
        for el in Venue1.find_parents("div",attrs = {"class":"items__txt__time"}):
            Venue = el.get_text().strip()
        Time1 = item.find ("a", attrs = {"class":"items__txt__time__icon"})
        for el in Time1.find_parents("div",attrs = {"class":"items__txt__time"}):
            Time = el.get_text().strip()
        Price = item.find ("div", attrs = {"class":"items__txt__price"}).find("span").text.replace("\n"," ").strip()
        Link = item.find("div",attrs = {"class":"items__txt__title"}).find("a").get("href")
        data.append((Picture,Title,Venue,Time,Price,Link))
    searchBtn = driver.find_element_by_class_name('btn-next')
    searchBtn.click()
    time.sleep(10)
driver.quit()
df = pd.DataFrame(data, columns = header)
df.to_csv('Dramas.csv',sep=',', encoding = 'utf_8_sig')
df
```
